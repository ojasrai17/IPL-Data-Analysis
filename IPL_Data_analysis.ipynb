```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
sns.set_palette("Set2")

```

```python
# Load CSV
matches_df = pd.read_csv('matches.csv')

# Load Excel files
auction_df = pd.read_excel('/content/Auction.xlsx')
batsmen_df = pd.read_excel('/content/Top_100_batsman.xlsx')
bowlers_df = pd.read_excel('/content/Top_100_bowlers.xlsx')
```

```python
print("Matches Data:")
display(matches_df.head())

print("Auction Data:")
display(auction_df.head())

print("Top 100 Batsmen:")
display(batsmen_df.head())

print("Top 100 Bowlers:")
display(bowlers_df.head())
```

```python
matches_df['winner'].value_counts().head(10).plot(kind='bar', title="Most Match Wins", figsize=(10,5))
plt.xlabel("Team")
plt.ylabel("Wins")
plt.show()
```

```python
# Make sure to use the correct column name
top_paid = auction_df.sort_values(by='price(in_crore_indian_rupees)_', ascending=False).head(10)

# Plotting the Top 10 Most Expensive Players
sns.barplot(x='price(in_crore_indian_rupees)_', y='player', data=top_paid)
plt.title("Top 10 Most Expensive Players")
plt.xlabel("Price (in Crores)")
plt.ylabel("Player")
plt.show()
```

```python
# Sort batsmen by 'Runs' in descending order and select top 10
top_batsmen = batsmen_df.sort_values(by='Runs', ascending=False).head(10)

# Plotting the Top 10 Run Scorers
sns.barplot(x='Runs', y='PLAYER', data=top_batsmen)
plt.title("Top 10 Run Scorers")
plt.xlabel("Runs")
plt.ylabel("Player")
plt.show()
```

```python
# Sort bowlers by 'Wkts' in descending order and select top 10
top_bowlers = bowlers_df.sort_values(by='Wkts', ascending=False).head(10)

# Plotting the Top 10 Wicket Takers
sns.barplot(x='Wkts', y='PLAYER', data=top_bowlers)
plt.title("Top 10 Wicket Takers")
plt.xlabel("Wickets")
plt.ylabel("Player")
plt.show()
```
